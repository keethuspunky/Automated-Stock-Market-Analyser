```python
def download_stock_data(symbol="AAPL", start="2015-01-01", end="2024-12-31"):
    df = yf.download(symbol, start=start, end=end, progress=False)

    # Keep required columns
    df = df[["Open", "High", "Low", "Close", "Volume"]]

    # Clean data
    df.dropna(inplace=True)

    return df
```

```python
# Download data
df = download_stock_data("AAPL")

# Save dataset
df.to_csv("AAPL_stock_data.csv")

print("Dataset saved as AAPL_stock_data.csv")
df.head()
```

```python
def SMA(series, window):
    return series.rolling(window).mean()
```

```python
def RSI(series, period=14):
    series = series.squeeze()  # ensures 1D

    delta = series.diff()

    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)

    avg_gain = gain.rolling(period).mean()
    avg_loss = loss.rolling(period).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))

    return rsi
```

```python
def breakout_levels(series, window=20):
    high_break = series.rolling(window).max()
    low_break = series.rolling(window).min()
    return high_break, low_break
```

```python
df["SMA_10"] = SMA(df["Close"], 10)
df["SMA_50"] = SMA(df["Close"], 50)
df["RSI_14"] = RSI(df["Close"], 14)

df["Break_High_20"], df["Break_Low_20"] = breakout_levels(df["Close"], 20)
```

```python
def backtest_sma_strategy(df):
    price = df["Close"]

    signal = (df["SMA_10"] > df["SMA_50"]).astype(int)

    # Avoid look-ahead bias
    position = signal.shift(1).fillna(0)

    returns = price.pct_change().fillna(0)

    strategy_returns = position * returns
    equity_curve = (1 + strategy_returns).cumprod()

    return strategy_returns, equity_curve
```

```python
strategy_returns, equity_curve = backtest_sma_strategy(df)

equity_curve.tail()
```

```python
total_return = equity_curve.iloc[-1] - 1
annualized_return = (equity_curve.iloc[-1]) ** (252 / len(equity_curve)) - 1

print("Total Return:", round(total_return * 100, 2), "%")
print("Annualized Return:", round(annualized_return * 100, 2), "%")
```

```python
def permutation_test(strategy_returns, n_perm=1000):
    observed_mean = strategy_returns.mean()
    perm_means = []

    for _ in range(n_perm):
        shuffled = np.random.permutation(strategy_returns)
        perm_means.append(shuffled.mean())

    perm_means = np.array(perm_means)
    p_value = np.mean(perm_means >= observed_mean)

    return p_value
```

```python
p_value = permutation_test(strategy_returns.values)
print("Permutation test p-value:", p_value)
```
